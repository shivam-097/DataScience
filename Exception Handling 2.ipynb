#### 2. Write a python program to print Python Exception Hierarchy

```python
'''Ans.'''

def Class_hier(cls, i=0):
    print('-'*i, cls.__name__)
    for j in cls.__subclasses__():
        Class_hier(j, i+2)

print("Hierarchy for exceptions is = ")

Class_hier(Exception)
```

#### 6. List down some best practices for exception handling

1. Always use specific exception classes as much as possible
2. Always mention descriptive error message so that user can understand the error better
3. Use logging for error logging
4. Wherever exception may occur use try and except block for its handling
5. Avoid multiple exception handling
6. Make proper documentation of all the exceptions encountered.
