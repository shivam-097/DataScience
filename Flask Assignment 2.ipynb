Q1. Explain GET and POST methods.

Ans. The two common method in HTTP are GET and POST.

* GET is used to request data from a specified resource. It is used when the data is not sensitive because the keyword entered gets displayed in URL.
* POST is used to send data to a server to create/ update a resource. Mostly used for authentication enabled application as now information is diplayed in the URL

Q2. Why is request used in Flask?

Ans. Request is an object that contains all the data sent from the client to server. Following are the attributes of the request objects
* Form =  It is the dictionary object which contains the key value pair of form parameters and their values.
* args = It is parsed from the URL. It is the part of the URL which is specified in the URL after question mark(?)
* Cookies = It is the dictionary object containing cookie names and values
* files = It contains the data related to the uploaded file
* method = It is the current request method (get or post)

Q3. Why is redirect() used in Flask?

Ans. The redirect() function is used in Flask to redirect the user to another URL with the provided status code.

Q4. What are templates in Flask? Why is the render_template() function used?

render_template() function is used to generate output from a template file.
